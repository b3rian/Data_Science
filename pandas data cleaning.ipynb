```python
xy = pd.read_excel(r"C:\Users\ADMIN\Downloads\Customer Call List.xlsx")
xy.drop_duplicates(inplace=True)
xy.drop(columns = 'Not_Useful_Column', inplace=True)
# xy['Last_Name'] = xy['Last_Name'].str.lstrip('...')
# xy['Last_Name'] = xy['Last_Name'].str.lstrip('/')
# xy['Last_Name'] = xy['Last_Name'].str.rstrip('_')
xy['Last_Name'] = xy['Last_Name'].str.strip('123./_')
xy['Phone_Number'] = xy['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True)
xy['Phone_Number'] = xy['Phone_Number'].apply(lambda x: str(x))
xy['Phone_Number'] = xy['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
xy['Phone_Number'] =xy['Phone_Number'].str.replace('nan--','')
xy['Phone_Number'] =xy['Phone_Number'].str.replace('Na--', '')
xy[['Street_Address', 'State', 'Zip_Code']]= xy['Address'].str.split(',',n = 2, expand=True)
xy.drop(columns = 'Address', inplace=True)
xy['Paying Customer']= xy['Paying Customer'].str.replace('Yes', 'Y')
xy['Paying Customer']= xy['Paying Customer'].str.replace('No', 'N')
xy['Do_Not_Contact'] = xy['Do_Not_Contact'].str.replace('Yes', 'Y')
xy['Do_Not_Contact'] = xy['Do_Not_Contact'].str.replace('No', 'N')
xy.fillna('', inplace=True)
xy.replace('N/a', '', inplace=True)

for x in xy.index:
    if xy.loc[x, 'Do_Not_Contact'] == 'Y':
        xy.drop(x, inplace=True)

for x in xy.index:
    if xy.loc[x, 'Phone_Number'] == '':
        xy.drop(x, inplace=True)


xy.reset_index(drop=True, inplace=True)
xy
```
