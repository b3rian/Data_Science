```python
np.random.seed(19680801)
data = {'a': np.arange(50),
'c': np.random.randint(0,50,50),
'd': np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(figsize = (5, 2.7), layout = 'constrained')
ax.plot(x, x, label = 'linear')
ax.plot(x, x**2, label = 'quadratic')
ax.plot(x, x**3, label = 'cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend()
```

```python
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 6))

# Define a 2x2 grid and place subplots
plt.subplot(2, 2, 1)  # First subplot (row=1, col=1)
plt.plot(x, np.sin(x))
plt.title("Sine Function")

plt.subplot(2, 2, 2)  # Second subplot (row=1, col=2)
plt.plot(x, np.cos(x))
plt.title("Cosine Function")

plt.subplot(2, 2, 3)  # Third subplot (row=2, col=1)
plt.plot(x, np.tan(x))
plt.title("Tangent Function")

plt.subplot(2, 2, 4)  # Fourth subplot (row=2, col=2)
plt.plot(x, np.exp(-x))
plt.title("Exponential Decay")

plt.tight_layout()  # Adjusts spacing
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-X**2 - Y**2)

fig, ax = plt.subplots()
contour = ax.contourf(X, Y, Z, cmap='coolwarm')
fig.colorbar(contour)

plt.title("Contour Plot with Colorbar")
plt.show()
```

```python
# Load dataset
diamonds = sns.load_dataset("diamonds")

# Create a scatter plot of price vs carat
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(diamonds["carat"], diamonds["price"], alpha=0.3)
plt.xscale('linear')
plt.yscale('linear')
plt.xlabel("Carat")
plt.ylabel("Price")
plt.title("Linear Scale")

plt.subplot(1, 2, 2)
plt.scatter(diamonds["carat"], diamonds["price"], alpha=0.3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Carat (log scale)")
plt.ylabel("Price (log scale)")
plt.title("Logarithmic Scale")

plt.tight_layout()
plt.show()
```

```python
# Load iris dataset
iris = sns.load_dataset("iris")
```

```python
# Define colors for each species
colors = {'setosa': 'red', 'versicolor': 'blue', 'virginica': 'green'}

# Create scatter plot
plt.figure(figsize=(8,6))

for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], 
                label=species, color=colors[species])

# Adding labels and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Sepal Dimensions")

# Add legend
plt.legend(title="Species", loc="upper right", fontsize=12, frameon=True)

# Show plot
plt.show()
```
