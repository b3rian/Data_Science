```python
new_data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Monitor', 'Keyboard'],
    'Price ($)': [1200, 800, 300, 250, 100],
    'Stock': [10, 50, 30, 20, 100],
    'Rating': [4.5, 4.7, 4.3, 4.1, 3.9]
}

dataset = pd.DataFrame(new_data)
```

```python
def highlight_low_stock(val):
    color = 'red' if val < 20 else 'black'
    return f'color: {color}'

dataset.style.map(highlight_low_stock, subset=['Stock'])

 
```
